# Embed data on Colab

```python
import pandas as pd
from sentence_transformers import SentenceTransformer
```

```python
# Set path to the file with the text chunks on your Google Drive.
df = pd.read_parquet("/content/drive/MyDrive/06_text_chunks.parq")
```

```python
model_path = "jinaai/jina-embeddings-v2-base-de"
model = SentenceTransformer(
    model_path,
    trust_remote_code=True,
)
print("Max Sequence Length:", model.max_seq_length)
model.max_seq_length = 512
```

```python
embeddings = model.encode(
    df.chunk_text.values,
    batch_size=128,
    convert_to_tensor=False,
    normalize_embeddings=True,
    show_progress_bar=True,
)
```

```python
df["embeddings"] = list(embeddings)
df.to_parquet("/content/drive/MyDrive/chunks_embedded.parq")
```
